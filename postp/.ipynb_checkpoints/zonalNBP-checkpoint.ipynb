```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import glob
import pandas as pd
```

```python
def amean(da):
    #annual mean
    m  = da['time.daysinmonth']
    xa = 1/365*(m*da).groupby('time.year').sum().compute()
    xa.name=da.name
    xa.attrs=da.attrs
    return xa

def gsum(da,la):
    x=(da*la).sum(dim=['lat','lon']).compute()
    return x
```

```python
dhist='/glade/campaign/cgd/tss/projects/TRENDY/v12/{}/lnd/hist/'
dproc='/glade/campaign/cgd/tss/projects/TRENDY/v12/{}/lnd/proc/tseries/month_1/'
dtrendy='/glade/campaign/cgd/tss/projects/TRENDY/v12/{}/lnd/proc/trendy/'
```

### write a land area netcdf for each simulation
 - though they should all be the same

```python
for sim in ['S0','S1','S2','S3']:
    f=sorted(glob.glob(dhist.format(sim)+'*h0*'))[0]
    ds=xr.open_dataset(f)
    area=1e6*ds.area
    area.attrs={'long_name':'grid cell areas','units':'m2'}
    la=xr.Dataset()
    la['area']=area
    la['landfrac']=ds.landfrac
    fout=(dtrendy+'CLM5.0_{}_area.nc').format(sim,sim)
    la.to_netcdf(fout)
```

```python
sim='S0'
f=glob.glob(dtrendy.format(sim)+'*_area.nc')[0]
print(f)
tmp=xr.open_dataset(f)
laglobal=tmp.area*tmp.landfrac
lat=laglobal.lat
la={'Global':laglobal,
    'North':laglobal.where(lat>30),
    'Tropics':laglobal.where(abs(lat)<30),
    'South':laglobal.where(lat<-30)}
for i,region in enumerate(la):
    plt.subplot(2,2,i+1)
    la[region].plot(add_colorbar=False)
    plt.title(region)
    plt.ylabel('')
    plt.xlabel('')
    plt.xticks([])
    plt.yticks(-60+30*np.arange(5))
```

### process NBP a la TRENDY

```python
for sim in ['S0','S1','S2','S3']:
    f=glob.glob(dproc.format(sim)+'*h0.NBP.*')[0]
    ds=xr.open_dataset(f,chunks={'time':60})
    yr0=ds['time.year'][0].values
    ds['time']=xr.cftime_range(str(yr0),periods=len(ds.time),freq='MS',calendar='noleap')
    f=glob.glob(dtrendy.format(sim)+'*_area.nc')[0]
    tmp=xr.open_dataset(f)
    laglobal=tmp.area*tmp.landfrac
    lat=laglobal.lat
    la={'Global':laglobal,
        'North':laglobal.where(lat>30),
        'Tropics':laglobal.where(abs(lat)<30),
        'South':laglobal.where(lat<-30)}
    cf=1e-15*24*60*60*365 #PgC/yr
    xs={reg:cf*gsum(amean(ds.NBP),la[reg]) for reg in la}
    df=pd.DataFrame({region:xs[region] for region in xs},index=xs['Global'].year.values)
    df.index.name='Year'
    f=(dtrendy+'CLM5.0_{}_zonalNBP.txt').format(sim,sim)
    df.to_csv(f)
```

### plot results directly from csvs

```python
das=[]
sims=['S0','S1','S2','S3']
for sim in sims:
    f=(dtrendy+'CLM5.0_{}_zonalNBP.txt').format(sim,sim)
    df=pd.read_csv(f)
    das.append(xr.DataArray([df[reg].values for reg in la],dims=['region','year']))
nbp=xr.concat(das,dim='sim')
nbp['year']=df.Year.values
nbp['sim']=sims
nbp['region']=[r for r in la]
```

```python
region='Global'
plt.plot(nbp.year,nbp.sel(region=region).rolling({'year':20}).mean().T)
plt.legend(nbp.sim.values)
plt.ylabel('NBP (PgC/yr)\n20yr running mean')
plt.xlim([1700,2025])
plt.xlabel('year')
plt.title('TRENDYv12: CLM5.0 Global NBP')
plt.savefig('TRENDYv12_CLM5.0_NBP_global.png',dpi=300,bbox_inches='tight')
```

```python
for i,region in enumerate(nbp.region.values):
    plt.subplot(2,2,i+1)
    plt.plot(nbp.year,nbp.sel(region=region).cumsum(dim='year').T)
    plt.xlim([1700,2025])
    if i%2==0:
        plt.ylabel('Cumulative NBP (PgC)')
    plt.title(region)
    if i<2:
        plt.xticks(1700+100*np.arange(4),[])
    else:
        plt.xlabel('year')
plt.subplot(221)
plt.legend(nbp.sim.values)
plt.subplots_adjust(wspace=0.3)
plt.savefig('TRENDYv12_CLM5.0_NBPcumulative.png',dpi=300,bbox_inches='tight')
```
